```julia
using Pkg
Pkg.activate(".")
Pkg.add("Plots")
```

```julia
function H(n,x)
    if n==0 return 1 end
    if n==1 return 2x end
    return 2x .* H(n-1,x) .- ( 2(n-1) .* H(n-2,x) )
end
```

```julia
ψ₀(n,x) = (1.0/sqrt(2.0^n * factorial(n) * sqrt(pi))) * exp.(-x.^2 / 2) .* H(n,x)
```

```julia
x = range(-4,4,length=1_001)
```

```julia
using Plots

N = 0:15
anim = Animation()

plt = plot(ylim=(-0.1,0.75))
for n in N
    plot(plt, x, ψ₀(n,x) .* ψ₀(n,x)
        ,label="n = $(n)"
    )
    frame(anim)
end

gif(anim, "psi-plots.gif", fps = 2)
```
